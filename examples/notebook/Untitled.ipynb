```python
import numpy as np
import matplotlib.pyplot as plt


xlim = (-7, 7)
ylim = (-7, 7)
goal_positions = np.array(
            [
                [5, 0],
                [-5, 0],
                [0, 5],
                [0, -5]
            ],
            dtype=np.float32
)

# Multigoal
fig_env = plt.figure(figsize=(7, 7))
ax = fig_env.add_subplot(111)
ax.axis('equal')

delta = 0.01
x_min, x_max = tuple(1.1 * np.array(xlim))
y_min, y_max = tuple(1.1 * np.array(ylim))
X, Y = np.meshgrid(
    np.arange(x_min, x_max, delta),
    np.arange(y_min, y_max, delta)
)
goal_costs = np.amin([
    (X - goal_x) ** 2 + (Y - goal_y) ** 2
    for goal_x, goal_y in goal_positions
], axis=0)
costs = goal_costs

contours = ax.contour(X, Y, costs, 20)
ax.clabel(contours, inline=1, fontsize=10, fmt='%.0f')
ax.set_xlim([x_min, x_max])
ax.set_ylim([y_min, y_max])
goal = ax.plot(goal_positions[:, 0], goal_positions[:, 1], 'ro')

plt.show()
```
